```python
def reduce_dataset_size(dataset, target_size):
    reduced_dataset = ([], [])
    class_counts = {}

    images, labels = dataset[0], dataset[1]

    for image, label in zip(images, labels):
        if label not in class_counts:
            class_counts[label] = 0

        if class_counts[label] < target_size:
            reduced_dataset[0].append(image)
            reduced_dataset[1].append(label)
            class_counts[label] += 1

    # Convert to numpy arrays
    reduced_dataset = (np.array(reduced_dataset[0]), np.array(reduced_dataset[1]))

    return reduced_dataset
```

```python
# Reduce the size of the training dataset to 600 images per label
new_train_dataset = reduce_dataset_size(dataset[0], target_size=600)

# Reduce the size of the testing dataset to 100 images per label
new_test_dataset = reduce_dataset_size(dataset[1], target_size=100)

# Combine into a new dataset
new_dataset = (new_train_dataset, new_test_dataset)
```
